```python
# 라이브러리 import
import numpy as np
import os
import pandas as pd

import statsmodels.api as sm

from sklearn.metrics import mean_squared_error
```

```python
# 엑셀(xlsx) 데이터 파일 불러오기
data = pd.read_excel('wine_aroma.xlsx')
```

```python
print(data)
```

```python
type(data)
```

```python
# numpy 배열로 변환
data_np = data.to_numpy()
print(type(data_np))
# print(data_np)
```

```python
# 독립변수와 종속변수 분리
X = data_np[:,:-1]
y = data_np[:,-1]
```

```python
print(X)
```

```python
print(y)
```

```python
# 상수항 추가
X_aug = sm.add_constant(X, prepend=False)
```

```python
pd.DataFrame(X_aug)
```

```python
# 선형 회귀 모형 추정
reg = sm.OLS(y, X_aug).fit()
print(reg.summary())
```

```python
y_pred = reg.predict(X_aug)
```

```python
y_pred
```

```python
print(mean_squared_error(y, y_pred))
```
